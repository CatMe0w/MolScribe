```python
import pandas as pd

score_headers = ["graph_smiles", "graph_graph", "graph_chiral"]
score_dfs = []
for dataset, cur_eval_files in zip(datasets, zip(*eval_files)):
    eval_res = []
    for eval_file in cur_eval_files:
        with open(eval_file, 'r') as f:
            eval_res.append(json.load(f))
    scores = []
    for cur_eval_res in eval_res:
        score_row = []
        for score_header in score_headers:
            score_row.append(cur_eval_res[score_header])
        scores.append(score_row)
    score_dfs.append(pd.DataFrame(scores, columns=[f"{dataset}_{score_header}" for score_header in score_headers]))

pd.concat(score_dfs, axis=1)
```
