# Indigo and ChemDraw

```python
import pandas as pd

data = pd.read_csv('../data/molbank/indigo-data/valid.csv')
data_out = pd.read_csv('../data/molbank/chemdraw-data/valid.csv')
pred_in = pd.read_csv('../output/indigo/swin_base_20_dynamic_aug/prediction_valid_indigo.csv')
pred_out = pd.read_csv('../output/indigo/swin_base_20_dynamic_aug/prediction_valid_chemdraw.csv')
```

```python
import numpy as np

print(np.mean(data['SMILES'].values == pred_in['SMILES'].values))
print(np.mean(data['SMILES'].values == pred_out['SMILES'].values))

print(np.mean(data['InChI'].values == pred_in['SMILES_InChI'].values))
print(np.mean(data['InChI'].values == pred_out['SMILES_InChI'].values))
```

```python
exact_match_in = (data['InChI'].values == pred_in['SMILES_InChI'].values)
exact_match_out = (data['InChI'].values == pred_out['SMILES_InChI'].values)

stats = np.zeros((2,2))

for i, smiles in enumerate(data['SMILES'].values):
    stats[int(exact_match_in[i]), int(exact_match_out[i])] += 1

print(stats)
```

```python
import cv2
import matplotlib.pyplot as plt
import rdkit.Chem as Chem
from rdkit.Chem import Draw
from IPython.display import SVG

BASE = '../'

cnt = 0 
for i, row in data.iterrows():
    if exact_match_in[i] and not exact_match_out[i]:
        cnt += 1
        print(row['SMILES'])
        pred_smiles_in = pred_in.loc[i,'SMILES']
        pred_smiles_out = pred_out.loc[i,'SMILES']
        print('pred in:', pred_in.loc[i,'SMILES_InChI'])
        print('pred out:', pred_out.loc[i,'SMILES_InChI'])
        image_in = cv2.imread(BASE + row['file_path'])
        image_out = cv2.imread(BASE + data_out.loc[i,'file_path'])
        plt.figure(figsize=(10,5))
        plt.subplot(1,2,1)
        plt.imshow(image_in)
        plt.subplot(1,2,2)
        plt.imshow(image_out)
        plt.show()
        mol_in = Chem.MolFromSmiles(pred_smiles_in)
        mol_out = Chem.MolFromSmiles(pred_smiles_out)
        img = Draw.MolsToGridImage([mol_in,mol_out], subImgSize=(300,300), molsPerRow=2, useSVG=True)
        display((img))
    if cnt == 10:
        break
```

```python
# print(np.mean(data['smiles'].values == pred['smiles'].values))
print(np.mean(data['InChI'].values == pred['SMILES_InChI'].values))

def normalize_smiles(smiles):
    try:
        mol = Chem.MolFromSmiles(smiles)
        smiles = Chem.MolToSmiles(mol)
        return smiles
    except:
        return ""

gold_smiles = [normalize_smiles(s) for s in data['SMILES'].values]
pred_smiles = [normalize_smiles(s) for s in pred['SMILES'].values]
print(np.mean(np.array(gold_smiles) == np.array(pred_smiles)))
```

```python
import albumentations as A

transform = A.ToGray(p=1)

exact_match = (data['InChI'].values == pred['SMILES_InChI'].values)

BASE = '../'
# BASE = '../data/molbank/real-acs-evaluation/'

cnt = 0 
for i, row in data.iterrows():
    if not exact_match[i]:
        cnt += 1
        print(row['file_path'])
        pred_smiles = pred.loc[i,'SMILES']
        pred_inchi = pred.loc[i,'SMILES_InChI']
        print('gold:', row['SMILES'])
        print('pred:', pred_smiles)
        print('gold:', row['InChI'])
        print('pred:', pred_inchi)
        image = cv2.imread(BASE + row['file_path'])
        plt.figure(figsize=(8,4))
        plt.subplot(1,2,1)
        plt.axis('off')
        plt.imshow(image)
#         plt.subplot(1,2,2)
#         plt.axis('off')
#         transformed = transform(image=image)['image']
#         plt.imshow(transformed)
        plt.show()
        mol = Chem.MolFromSmiles(pred_smiles)
        mol1 = Chem.MolFromSmiles(row['SMILES'])
        img = Draw.MolsToGridImage([mol,mol1], subImgSize=(250,250), molsPerRow=2, useSVG=True)
        display(img)
#     if cnt == 50:
#         break
```
