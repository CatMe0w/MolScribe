```python
mol = indigo.loadMoleculeFromFile('../data/molbank/tmp/mol.mol')

xx, yy = [], []
for atom in mol.iterateAtoms():
    v = atom.xyz()
    xx.append(v[0])
    yy.append(-v[1])

xx = np.array(xx)
xx = (xx - min(xx)) / (max(xx) - min(xx))
yy = np.array(yy)
yy = (yy - min(yy)) / (max(yy) - min(yy))

xx = 37 + xx * (451-37)
yy = 22 + yy * (325-22)
plt.scatter(xx, yy)
plt.show()
print([[round(xx[i]), round(yy[i])] for i in range(len(xx))])

n = len(xx)
edges = np.zeros((n, n), dtype=int)
for bond in mol.iterateBonds():
    s = bond.source().index()
    t = bond.destination().index()
    # 1/2/3/4 : single/double/triple/aromatic
    edges[s, t] = bond.bondOrder()
    edges[t, s] = bond.bondOrder()
    if bond.bondStereo() in [5, 6]:
        edges[s, t] = bond.bondStereo()
        edges[t, s] = 11 - bond.bondStereo()
print(edges.tolist())
```
