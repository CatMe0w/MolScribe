```python
import pandas as pd

ROOT_PATH = "/data/rsg/nlp/yujieq/bms/"

gold = pd.read_csv(ROOT_PATH + 'data/valid_folds.csv')
pred = pd.read_csv(ROOT_PATH + 'output/resnet50_input_320_atomtok_epoch_12/prediction_valid.csv')
```

```python
gold
```

```python
smiles = gold['SMILES_atomtok'].values
print(max([len(s.split()) for s in smiles]))
```

```python
image_ids = gold['image_id'].values
gold_smiles = gold['SMILES'].values
pred_smiles = pred['SMILES'].values
gold_inchi = gold['InChI'].values
pred_inchi = pred['InChI'].values

errors = []
for i in range(len(image_ids)):
    if gold_inchi[i] != pred_inchi[i]:
        errors.append(i)

print(len(errors), '/', len(image_ids))
```

```python
import Levenshtein
import matplotlib.pyplot as plt

d_list = []
for idx in errors:
    gs = gold_smiles[idx]
    ps = pred_smiles[idx]
    distance = Levenshtein.distance(gs, ps)
    d_list.append(distance)

plt.figure()
plt.hist(d_list,bins=50)
plt.show()

d_list = []
for idx in errors:
    gi = gold_inchi[idx]
    pi = pred_inchi[idx]
    distance = Levenshtein.distance(gi, pi)
    d_list.append(distance)

plt.figure()
plt.hist(d_list,bins=50)
plt.show()
```

```python
from collections import Counter
print(mol.GetNumAtoms(), mol.GetNumBonds(), mol.GetNumHeavyAtoms())
print(mol.GetRingInfo().NumRings())
atoms = mol.GetAtoms()
for atom in atoms:
    print(atom.GetSymbol(), atom.GetSmarts(), atom.GetChiralTag())
chiral = [atom.GetChiralTag() for atom in mol.GetAtoms()]
print(Counter(chiral))
print('-'*10)
bonds = mol.GetBonds()
for bond in bonds:
    print(bond.GetBeginAtom().GetSmarts(), bond.GetSmarts(), bond.GetEndAtom().GetSmarts())
```

```python
from rdkit.Chem import rdDepictor
from rdkit.Chem.Draw import rdMolDraw2D
def moltosvg(mol,molSize=(500,300),kekulize=True):
    mc = Chem.Mol(mol.ToBinary())
    if not mc.GetNumConformers():
        rdDepictor.Compute2DCoords(mc)
    drawer = rdMolDraw2D.MolDraw2DSVG(molSize[0],molSize[1])
    drawer.DrawMolecule(mc)
    drawer.FinishDrawing()
    svg = drawer.GetDrawingText()
    # It seems that the svg renderer used doesn't quite hit the spec.
    # Here are some fixes to make it work in the notebook, although I think
    # the underlying issue needs to be resolved at the generation step
    return svg.replace('svg:','')
```

```python
from matplotlib import pyplot as plt
from IPython.display import SVG
import cv2
import rdkit
import rdkit.Chem as Chem
from rdkit.Chem import AllChem
from rdkit.Chem import Draw

def get_train_file_path(image_id):
    return ROOT_PATH + "data/train/{}/{}/{}/{}.png".format(image_id[0], image_id[1], image_id[2], image_id)

for idx in errors[:20]:
    image_id = image_ids[idx]
    gs = gold_smiles[idx]
    ps = pred_smiles[idx]
    gi = gold_inchi[idx]
    pi = pred_inchi[idx]
    print(image_id)
    print("## SMILES ##")
    print('gold', gs)
    print('pred', ps)
    print('distance', Levenshtein.distance(gs, ps))
    print("## InChI ##")
    print('gold', gi)
    print('pred', pi)
    print('distance', Levenshtein.distance(gi, pi))
    file_path = get_train_file_path(image_id)
    image = cv2.imread(file_path)
    w, h, _ = image.shape
    plt.figure(figsize=(w/10, h/10))
    plt.imshow(image)
    plt.show()
    mol_gold = Chem.MolFromSmiles(gs)
#     display(SVG(moltosvg(mol_gold)))
    mol_pred = Chem.MolFromSmiles(ps)
#     display(SVG(moltosvg(mol_pred)))
    img = Draw.MolsToGridImage([mol_gold,mol_pred], subImgSize=(400,400), molsPerRow=2, useSVG=True)
    display(SVG(img))
```
