# LD by molecule size

```python
import Levenshtein

def compute_ld(gold_list, pred_list):
    result = []
    for gs, ps in zip(gold_list, pred_list):
        result.append(Levenshtein.distance(gs, ps))
    return result

pred['SMILES_LD'] = compute_ld(gold['SMILES'].values, pred['SMILES'].values)
pred['InChI_LD'] = compute_ld(gold['InChI'].values, pred['InChI'].values)
pred['merge_InChI_LD'] = compute_ld(gold['InChI'].values, pred['merge_InChI'].values)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

size = []
count = []
inchi_ld = []
for i in range(9):
    size.append(str(i*10))
    indices = (i*10 < gold['num_atom']) & (gold['num_atom'] <= (i+1)*10)
    print(indices.sum())
    count.append(indices.sum())
    metric = pred[['SMILES_LD','InChI_LD','merge_InChI_LD']][indices].mean()
    print(metric)
    inchi_ld.append(metric['merge_InChI_LD'])

df = pd.DataFrame({'size':size, 'count':count, 'inchi_ld':inchi_ld})

fig, ax1 = plt.subplots()
color = 'tab:blue'
sns.barplot(x='size', y='inchi_ld', data=df, color=color)
ax1.tick_params(axis='y', labelcolor=color)
ax1.set_ylabel('inchi_ld', color=color)
ax2 = plt.twinx()
color = 'tab:orange'
sns.lineplot(x='size', y='count', data=df, color=color, marker='o', ax=ax2)
ax2.tick_params(axis='y', labelcolor=color)
ax2.set_ylabel('count', color=color)
```

```python
from IPython.display import SVG
import rdkit
import rdkit.Chem as Chem

def find_closest(smiles, smiles_list):
    min_d = len(smiles)
    min_smiles = None
    for s in smiles_list:
        if abs(len(smiles)-len(s)) > min_d:
            continue
        d = Levenshtein.distance(smiles, s)
        if d < min_d:
            min_d = d
            min_smiles = s
    return min_smiles, min_d

train_smiles = train['SMILES'].values
for smiles in gold['SMILES'].values[:10]:
    min_smiles, min_d = find_closest(smiles, train_smiles)
    print(smiles)
    print(min_smiles)
    print("Distance", min_d)
    mol_0 = Chem.MolFromSmiles(min_smiles)
    mol_1 = Chem.MolFromSmiles(smiles)
    img = Draw.MolsToGridImage([mol_0,mol_1], subImgSize=(400,400), molsPerRow=2, useSVG=True)
    display(SVG(img))
```

```python
image_ids = gold['image_id'].values
gold_smiles = gold['SMILES'].values
pred_smiles = pred['SMILES'].values
gold_inchi = gold['InChI'].values
pred_inchi = pred['merge_InChI'].values

errors = []
for i in range(len(image_ids)):
    if gold_smiles[i] != pred_smiles[i] and gold_inchi[i] != pred_inchi[i]:
        errors.append(i)

print(len(errors), '/', len(image_ids))
```

```python
import Levenshtein
import matplotlib.pyplot as plt
import collections

d_list = []
ds_to_di = collections.defaultdict(list)
for idx in errors:
    gs = gold_smiles[idx]
    ps = pred_smiles[idx]
    distance = Levenshtein.distance(gs, ps)
    d_list.append(distance)
    gi = gold_inchi[idx]
    pi = pred_inchi[idx]
    di = Levenshtein.distance(gi, pi)
    ds_to_di[distance].append(di)

plt.figure()
sns.histplot(d_list,bins=50)
plt.xlabel('SMILES_LD')
plt.show()
print(collections.Counter(d_list))

# print(len(ds_to_di[1]), sum([x<5 for x in ds_to_di[1]]))
plt.figure()
sns.histplot(ds_to_di[1],bins=50)
plt.xlabel('InChI_LD')
plt.show()

# d_list = []
# for idx in errors:
#     gi = gold_inchi[idx]
#     pi = pred_inchi[idx]
#     distance = Levenshtein.distance(gi, pi)
#     d_list.append(distance)

# plt.figure()
# plt.hist(d_list,bins=50)
# plt.show()
```

```python
from rdkit.Chem import rdDepictor
from rdkit.Chem.Draw import rdMolDraw2D
def moltosvg(mol,molSize=(500,300),kekulize=True):
    mc = Chem.Mol(mol.ToBinary())
    if not mc.GetNumConformers():
        rdDepictor.Compute2DCoords(mc)
    drawer = rdMolDraw2D.MolDraw2DSVG(molSize[0],molSize[1])
    drawer.DrawMolecule(mc)
    drawer.FinishDrawing()
    svg = drawer.GetDrawingText()
    # It seems that the svg renderer used doesn't quite hit the spec.
    # Here are some fixes to make it work in the notebook, although I think
    # the underlying issue needs to be resolved at the generation step
    return svg.replace('svg:','')
```
