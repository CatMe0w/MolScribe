# Compare examples

Compare old model and new model by displaying the predictions the new model fixed and those the new model broke.
There's also an option of only displaying predictions that contain "new symbols", i.e., tokens that aren't in the original vocab (`vocab_uspto.new.json`).

```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append('../')
```

```python
import pandas as pd
import json
import cv2
import matplotlib.pyplot as plt
import rdkit.Chem as Chem
from rdkit.Chem import Draw
from bms.chemistry import SmilesEvaluator

with open("../bms/vocab_uspto_new.json", 'r') as f:
    old_vocab = json.load(f)

gold_files = ['../data/molbank/Img2Mol/CLEF.csv',
              '../data/molbank/Img2Mol/JPO.csv',
              '../data/molbank/Img2Mol/UOB.csv',
              '../data/molbank/Img2Mol/USPTO.csv',
              '../data/molbank/Img2Mol/staker.csv',
              '../data/molbank/real-acs-evaluation/acs.csv'
             ]
# pred_files = ['../output/uspto/swin_base_aux_200k_new1/prediction_CLEF.csv',
#               '../output/uspto/swin_base_aux_200k_new1/prediction_JPO.csv',
#               '../output/uspto/swin_base_aux_200k_new1/prediction_UOB.csv',
#               '../output/uspto/swin_base_aux_200k_new1/prediction_USPTO.csv',
#               '../output/uspto/swin_base_aux_200k_new1/prediction_staker.csv',
#               '../output/uspto/swin_base_aux_200k_new1/prediction_acs.csv'
#              ]
pred_files = ['../output/uspto/swin_base_aux_200k_new1_augdata/prediction_CLEF.csv',
              '../output/uspto/swin_base_aux_200k_new1_augdata/prediction_JPO.csv',
              '../output/uspto/swin_base_aux_200k_new1_augdata/prediction_UOB.csv',
              '../output/uspto/swin_base_aux_200k_new1_augdata/prediction_USPTO.csv',
              '../output/uspto/swin_base_aux_200k_new1_augdata/prediction_staker.csv',
              '../output/uspto/swin_base_aux_200k_new1_augdata/prediction_acs.csv'
             ]
pred_header = 'post_SMILES'
pred_files_new = ['../output/uspto/swin_base_aux_200k_new1_augdata1/prediction_CLEF.csv_refactor',
              '../output/uspto/swin_base_aux_200k_new1_augdata1/prediction_JPO.csv_refactor',
              '../output/uspto/swin_base_aux_200k_new1_augdata1/prediction_UOB.csv_refactor',
              '../output/uspto/swin_base_aux_200k_new1_augdata1/prediction_USPTO.csv_refactor',
              '../output/uspto/swin_base_aux_200k_new1_augdata1/prediction_staker.csv_refactor',
              '../output/uspto/swin_base_aux_200k_new1_augdata1/prediction_acs.csv_refactor'
             ]
pred_header_new = 'post_SMILES'
dataset_indices = [5]

compare_new_symb = False
show_broke = True
show_fixed = False

indices = None
# indices = [(0, 871)]

display_details = True
count = 0
max_count = None
for dataset_idx, (gold_file, pred_file, pred_file_new) in enumerate(zip(gold_files, pred_files, pred_files_new)):
    if dataset_idx not in dataset_indices:
        continue
    gold_df = pd.read_csv(gold_file)
    pred_df = pd.read_csv(pred_file)
    pred_df_new = pd.read_csv(pred_file_new)
    assert all(gold_df.index == pred_df.index)
    
    evaluator = SmilesEvaluator(gold_df['SMILES'])
    eval_lst = evaluator.evaluate(pred_df[pred_header], True)['canon_smiles_details']
    eval_lst_new = evaluator.evaluate(pred_df_new[pred_header_new], True)['canon_smiles_details']
    
    for (i_g, row_g), (i_p, row_p), (i_p_new, row_p_new), eval_res, eval_res_new \
            in zip(gold_df.iterrows(), pred_df.iterrows(), pred_df_new.iterrows(), eval_lst, eval_lst_new):
        if indices is not None and (dataset_idx, i_g) not in indices:
            continue
        draw_graph = False
        symbols = eval(row_p['node_symbols'])
        symbols_new = eval(row_p_new['node_symbols'])
        
        if show_broke:
            if (not eval_res_new and eval_res):
                draw_graph = True
                if display_details:
                    print("NEW BROKE:")
                    print(f"Old {pred_header}:", row_p[pred_header])
                    print(f"New {pred_header_new}:", row_p_new[pred_header_new])
        
        if show_fixed:
            if (eval_res_new and not eval_res):
                draw_graph = True
                if display_details:
                    print("NEW FIXED:")
                    print(f"Old {pred_header}:", row_p[pred_header])
                    print(f"New {pred_header_new}:", row_p_new[pred_header_new])
        
        if compare_new_symb:
            for symbol in symbols:
                if symbol not in old_vocab and Chem.AtomFromSmiles(symbol) is None:
                    if (eval_res_new and not eval_res) or (not eval_res_new and eval_res):
                        draw_graph = True
                        if display_details:
                            if (eval_res_new and not eval_res):
                                print("NEW FIXED:")
                            else:
                                print("NEW BROKE:")
                            print("Old graph:", row_p['old_graph_SMILES'])
                            print("New graph:", row_p['graph_SMILES'])
                    break
                    
        if draw_graph:
            count += 1
            if display_details:
                print(i_g)
                print(row_g['image_id'])
                print(row_g['file_path'])
                print("Gold:", row_g['SMILES'])
                print("Symbols:", symbols)
                print("New symbols:", symbols_new)

                # PLOT ORIGINAL
                print("ORIGINAL:")
                file_path = '../' + row_g['file_path']
                img = cv2.imread(file_path)
                plt.figure(figsize=(5,5))
                plt.imshow(img)
                plt.show()

                print("GOLD SMILES:")
                mol = Chem.MolFromSmiles(row_g['SMILES'])
                display(Draw.MolToImage(mol))
                print("OLD                            NEW")
                mol1 = Chem.MolFromSmiles(row_p[pred_header])
                mol2 = Chem.MolFromSmiles(row_p_new[pred_header_new])
                try:
                    display(Draw.MolsToImage([mol1, mol2]))
                except:
                    pass
            
        if max_count is not None and count >= max_count:
            break
    if max_count is not None and count >= max_count:
        break

print(count)
```
